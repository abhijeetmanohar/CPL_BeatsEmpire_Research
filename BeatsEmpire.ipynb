```
#Sept 7
df_currentCash = df_groupbyPlayer[['currentCash']]
#df_currentCash

# player15_cash = df_currentCash.iloc[15]["currentCash"]
# max(player15_cash)

# Function to calculate the max cash value in a list
def maxCash(x):
  return(max(x))

# New Dataframe with playerUniqueId and currentCash
df_currentCash.loc[:,'currentCash'] = df_currentCash.loc[:,'currentCash'].apply(maxCash)
#df_currentCash

# Adding new column called 'cash>1M' with values 0 and 1 based on current Cash
label = []
for row in df_currentCash['currentCash']:
  if row > 1000000:
    label.append(1)
  else:
    label.append(0)

df_currentCash['cash_label'] = label
#df_currentCash
```

```
# Concatenating Feature Variables - Probability Factor data frame & Target variable data Frame

df_DecisionTree = pd.merge(df_currentCash, allPlayersTransitionsDF, left_index=True, right_index=True)
df_DecisionTree = df_DecisionTree.fillna(0) # VVVI : Replaced all NAN values with 0 
df_DecisionTree
```

```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

feature_cols = list(allPlayersTransitionsDF.columns) 
X = df_DecisionTree[feature_cols] # Features
y = df_DecisionTree.cash_label # Target variable

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy")

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```
# CREATING A MARKOV CHAIN

import numpy as np
class MarkovChain(object) : 
  # This function initializes the transition probabilities dict and the states dict to cultivate a markov chain. 
  def __init__(self, transition_prob) :
    """
    Initializing a Markov Instance. 
    Parameters -
    transition_prob : dict
       A dict object that takes a transition probabilities in Markov Chain
    """
    self.transition_prob = transition_prob
    self.states = list(transition_prob.keys())

  # This function generates the next state of the system, by considering the current system.
  def next_state(self, current_state) : 
    """
    Returns the next state based on the passed current state as parameter. 
    Parameters -
    current_state : str
      This is the current state of the system. 
    """
    return np.random.choice(
        self.states,
        p = [self.transition_prob[current_state][next_state] for next_state in self.states]
    )
  
  # This function generates the future states of the system, by considering the current system.
  def generate_states(self, current_state, numStates=10) :
    """
    Generates the next states of the system
    Parameters - 
    current_state : str 
      This is the current state of the system. 
    numStates : int
      This is the number of future states to generate. 
    """
    future_states = []
    for i in range(numStates) :
      next_state = self.next_state(current_state)
      future_states.append(next_state)
      current_state = next_state
    return future_states
```
